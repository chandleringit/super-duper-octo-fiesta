```python
# 取得資料get_salary程式
# 0811把最後一欄total人名會出現薪資移除

import time
import urllib.request as req
import urllib.error as err
import bs4 as bs

def get_salary(url):

    while True:
        try:
            resp = req.urlopen(url)
            break
        except err.HTTPError:
            print("網址失敗")
            time.sleep(1)
  # for saving
    table = []
    content = resp.read()
    html = bs.BeautifulSoup(content)
    target = html.find_all("table", {"class":"bserqJ__bserqJ _2gszxk__2gszxk o6j80t__o6j80t"})

    for r in target:
    # 找出所有名字欄
      names = r.find_all("td", {"class": "vTd-Ji__vTd-Ji"})
      # 找出所有薪資欄
      salaries = r.find_all("td", {"class": "RLrCiX__RLrCiX"})

      # 確保數量對齊，zip 讓名字和薪資一一配對
      for name, salary in zip(names, salaries):
          name_text = name.text.strip()
          salary_text = salary.text.strip()
          row ={"name":name_text,
         "salary":salary_text}

          table.append(row)
    table.pop()
    return table
```
